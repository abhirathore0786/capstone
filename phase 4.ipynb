```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\Hp\Downloads\outpt_3.8.2.csv')
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for state, group in df.groupby('STATE'):
    for district, data in group.groupby('DISTRICT'):
        ax.plot(data['YEAR'], data['MURDER'], label=f"{state} - {district}")

ax.set_xlabel('Year')
ax.set_ylabel('Murders')
ax.set_title('Murders by State and District over the Years')
ax.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='YEAR', y='MURDER', hue='STATE', data=df)
plt.xlabel('Year')
plt.ylabel('Murders')
plt.title('Murders by State over the Years')
plt.legend(title='State',loc='center left',bbox_to_anchor=(1, 0.5))

plt.show()
```
